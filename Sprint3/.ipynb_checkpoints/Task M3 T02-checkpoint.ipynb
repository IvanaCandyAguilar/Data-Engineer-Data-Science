+ We describe each column

|Column|Description|
|--------------|:--------------:|
|ID|Unique identifier code|
|Year|Four numerical digits, example 2008|
|Month|Two numerical digits, example 10|
|DayofMonth|Numerical digits, example 3|
|DayOfWeek|Days of week, 1-> Monday.....7->Sunday|
|DepTime| Actual departure time|
|CRSDepTime|Scheduled departure time|
|ArrTime|Actual arrive time|
|CRSArrTime|Scheduled arrive time|
|UniqueCarrier|Unique carrier code|
|FlightNum|**Flight number**|
|TailNum|**Plane tail number**|
|ActualElapsedTime|Actual elapsed time, in minutes|
|CRSElapsedTime|Scheduled elapsed time, in minutes|
|AirTime|Flight time in minutes |
|ArrDelay|Difference in minutes between scheduled and actual arrival time. Early arrivals show negative numbers, in minutes.|
|DepDelay|Difference in minutes between scheduled and actual departure time. Early departures show negative numbers, in minutes.|
|Origin|Origin airport code|
|Dest|Destination airport code|
|Distance|Distance between airports, in miles|
|TaxiIn|The time since wheels are down until arrival at destination airport gate, in minutes.|
|TaxiOut|The time elapsed between departure from the origin airport gate and wheels off, in minutes|
|Cancelled|If the flight was caancelled, 1 for yes, 0 for no|
|CancellationCode|Cancellation code, (Carrier, Weather, NAS, Security, Late airccraft)|
|Diverted|Diverted, 1 for yes, 0 for no|
|CarrierDelay|**Carrier delay, in minutes**|
|WeatherDelay|**Weather delay, in minutes**|
|NASDelay|**NAS delay, in minutes**|
|SecurityDelay|**Security delay, in minutes**|
|LateAircraftDelay|**Late aircraft, in minutes**|
