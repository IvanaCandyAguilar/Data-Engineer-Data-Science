```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.service import Service

try:
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Ejecutar en modo headless (sin interfaz gráfica)
    chrome_options.add_argument("--no-sandbox")  # Opción adicional para evitar errores en algunos sistemas
    
    # Especificar la ubicación del controlador Chrome WebDriver
    driver_path = 'D:/driver chrome/chromedriver.exe'  # Reemplaza con la ruta correcta a chromedriver
    
    # Crear un objeto Service con la ubicación del controlador Chrome WebDriver
    service = Service(driver_path)
    
    # Crear una instancia del controlador Chrome WebDriver
    driver = webdriver.Chrome(service=service, options=chrome_options)

    # Abrir el sitio web de Wikipedia en español
    #driver.get("https://es.wikipedia.org/wiki/Wikipedia:Portada")
    driver.get("https://es.wikipedia.org")

    # Esperar a que el formulario de búsqueda esté presente en la página
    search_form = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.NAME, "searchform")))

    # Hacer clic en el campo de búsqueda para asegurarse de que esté interactuable
    driver.execute_script("arguments[0].click();", search_form)

    # Ingresar la búsqueda utilizando execute_script
    driver.execute_script("arguments[0].value = 'universidad España';", search_form)

    # Enviar el formulario utilizando execute_script
    driver.execute_script("arguments[0].submit();", search_form)

    # Esperar a que los resultados estén presentes en la página
    results = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "mw-search-results")))

    # Imprimir el texto del primer resultado
    print(results[0].text)

    # Cerrar el navegador
    driver.quit()

except TimeoutException:
    print("No se encontraron elementos con la clase 'mw-search-results' dentro del tiempo especificado.")
```
