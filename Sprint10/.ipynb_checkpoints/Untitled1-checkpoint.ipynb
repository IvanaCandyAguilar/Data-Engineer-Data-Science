```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

# Configure Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (without GUI)
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Set path to chromedriver executable
chromedriver_path = "C:\Program Files (x86)\Google\chromedriver.exe"  # Replace with your actual path to chromedriver

# Create a new instance of the Chrome driver
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Open the webpage
driver.get("http://quotes.toscrape.com/")

# Wait for the results to load
try:
    results = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "quote"))
    )
    
    # Create an empty list to store the extracted data
    data = []
    
    
    # Process the job elements...
    for result in results:
        text = result.find_element(By.CLASS_NAME, "text").text
        author = result.find_element(By.CLASS_NAME, "author").text
        tags = result.find_element(By.CLASS_NAME, "tags").text
        
        # Append the extracted data to the list
        data.append({"text": text, "author": author, "tags": tags})
    
    # Create a dataframe from the extracted data
    df = pd.DataFrame(data)
    print(df)

except TimeoutException:
    print("Timeout occurred while waiting for the results to load.")

# Close the browser
driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

chrome_options = Options()
chrome_options.add_argument("--headless")  
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

chromedriver_path = "C:\Program Files (x86)\Google\chromedriver.exe"

service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

driver.get("https://es.wikipedia.org/w/index.php?search=universidad+Espa%C3%B1a&title=Especial:Buscar&profile=advanced&fulltext=1&ns0=1&ns100=1&ns104=1")

try:
    # Esperar a que los resultados estén presentes en la página
    results = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "mw-search-results")))

    # Crear listas vacías para almacenar los datos
    names = []
    descriptions = []
    updates = []

    # Extraer los datos de cada resultado
    for result in results:
        name_element = result.find_element(By.CLASS_NAME, "mw-search-result-heading")
        description_element = result.find_element(By.CLASS_NAME, "searchresult")
        update_element = result.find_element(By.CLASS_NAME, "mw-search-result-data")

        # Obtener el texto de cada elemento
        name = name_element.text
        description = description_element.text
        update = update_element.text

        # Agregar los datos a las listas
        names.append(name)
        descriptions.append(description)
        updates.append(update)

    # Crear el dataframe
    df = pd.DataFrame({"Name": names, "Description": descriptions, "Update": updates})
    print(df)

    # Cerrar el navegador
    driver.quit()

except TimeoutException:
    print("No se encontraron elementos con la clase 'mw-search-results' dentro del tiempo especificado.")
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

chromedriver_path = "C:\Program Files (x86)\Google\chromedriver.exe"

service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

driver.get("https://www.bolsasymercados.es/bme-exchange/es/Mercados-y-Cotizaciones/Acciones/Mercado-Continuo/Precios/ibex-35-ES0SI0000005")

try:
    # Esperar a que los resultados estén presentes en la página
    results = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".shares-table tr")))

    # Crear listas vacías para almacenar los datos
    data = []
    headers = ["Nombre", "Último", "% Dif.", "Máximo", "Mínimo", "Volumen", "Efectivo (miles €)", "Fecha", "Hora"]

    # Extraer los datos de cada fila de la tabla
    for row in results:
        cells = row.find_elements(By.TAG_NAME, "td")
        if len(cells) == len(headers):
            data.append([cell.text for cell in cells])

    # Crear el DataFrame
    df = pd.DataFrame(data, columns=headers)
    print(df)

    # Cerrar el navegador
    driver.quit()

except TimeoutException:
    print("No se encontraron elementos con la clase 'shares-table' dentro del tiempo especificado.")
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

chromedriver_path = "C:\Program Files (x86)\Google\chromedriver.exe"

service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

driver.get("https://es.wikipedia.org/w/index.php?search=universidad+Espa%C3%B1a&title=Especial:Buscar&profile=advanced&fulltext=1&ns0=1&ns100=1&ns104=1")

try:
    # Esperar a que los resultados estén presentes en la página
    results = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "mw-search-result"))
    )

    # Crear listas vacías para almacenar los datos
    names = []
    descriptions = []
    updates = []

    while True:
        # Extraer los datos de cada resultado
        for result in results:
            name_element = result.find_element(By.CLASS_NAME, "mw-search-result-heading")
            description_element = result.find_element(By.CLASS_NAME, "searchresult")
            update_element = result.find_element(By.CLASS_NAME, "mw-search-result-data")

            # Obtener el texto de cada elemento
            name = name_element.text
            description = description_element.text
            update = update_element.text

            # Verificar si la palabra inicial del nombre comienza con "universidad"
            if not name.startswith("universidad"):
                # Agregar los datos a las listas
                names.append(name)
                descriptions.append(description)
                updates.append(update)

        # Verificar si hay más páginas para visitar
        next_button = driver.find_element(By.CLASS_NAME, "mw-nextlink")
        if not next_button:
            break

        # Navegar a la siguiente página
        next_button.click()
        time.sleep(2)  # Esperar a que se cargue la página siguiente

        # Actualizar la lista de resultados
        results = driver.find_elements(By.CLASS_NAME, "mw-search-result")

    # Crear el dataframe
    df = pd.DataFrame({"Name": names, "Description": descriptions, "Update": updates})
    print(df.head(10))
    #df.to_csv('universities of Spain.csv', index=False)

    # Cerrar el navegador
    driver.quit()

except TimeoutException:
    print("No se encontraron elementos con la clase 'mw-search-result' dentro del tiempo especificado.")
```
