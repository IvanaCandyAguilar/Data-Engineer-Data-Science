```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Configurar Selenium
chrome_options = Options()
chrome_options.add_argument("--headless")  # Ejecutar en modo sin cabeza (sin ventana del navegador)
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chromedriver_path = "C:\Program Files (x86)\Google\chromedriver.exe"
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# URL de la página a hacer scraping
url = "https://www.promofarma.com/es/search?c=Salud&q=ortopedia"

# Acceder a la página
driver.get(url)

try:
    # Esperar a que el contenedor principal esté presente
    container = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "search-list"))
    )

    # Encontrar todos los productos en el contenedor
    products = container.find_elements(By.CLASS_NAME, "TextContent--small")

    # Crear listas para almacenar los datos
    descriptions = []
    prices = []

    # Iterar sobre los productos y obtener la descripción y el precio
    for product in products[:10]:
        description_element = product.find_element(By.CSS_SELECTOR, "h3.TextContent--small")
        price_element = product.find_element(By.CSS_SELECTOR, "span#price")

        description = description_element.text
        price = price_element.text

        descriptions.append(description)
        prices.append(price)

    # Crear el dataframe
    data = {"Description": descriptions, "Price": prices}
    df = pd.DataFrame(data)
    print(df.head(10))

    # Guardar en un archivo CSV
    df.to_csv("productos.csv", index=False)

finally:
    # Cerrar el navegador
    driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Configurar Selenium
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chromedriver_path = "C:\Program Files (x86)\Google\chromedriver.exe"
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# URL de la página a hacer scraping
url = "https://jobrun.dev/ofertas"

# Acceder a la página
driver.get(url)

try:
    # Esperar a que aparezca el contenedor de trabajos
    container = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "__wab_flex-container"))
    )

    # Buscar todos los elementos con la clase "__wab_flex-container"
    elements = container.find_elements(By.CLASS_NAME, "__wab_flex-container")

    # Obtener el nombre de la empresa de cada elemento
    empresas = []
    for element in elements:
        empresa_element = element.find_element(By.CLASS_NAME, "__wab_slot-string-wrapper")
        empresa = empresa_element.text
        empresas.append(empresa)

    # Obtener nombres únicos
    empresas_unicas = list(set(empresas))

    # Crear el dataframe
    data = {"Empresa": empresas_unicas}
    df = pd.DataFrame(data)
    print(df.head(10))

finally:
    # Cerrar el navegador
    driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Configurar Selenium
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chromedriver_path = "C:\Program Files (x86)\Google\chromedriver.exe"
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# URL de la página a hacer scraping
url = "https://jobrun.dev/ofertas"

# Acceder a la página
driver.get(url)

try:
    # Esperar a que aparezca el contenedor de trabajos
    container = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "plasmic_job_run_all__6OLuV.plasmic_job_run_root_reset__mR_qs.plasmic_job_run_plasmic_default_styles__lAHYy.plasmic_job_run_plasmic_tokens__GYzbj.PlasmicOfferGrid_root__b6R6l"))
    )

    # Buscar todos los elementos con la clase "__wab_slot-string-wrapper"
    elements = container.find_elements(By.CLASS_NAME, "__wab_slot-string-wrapper")

    # Estructurar los datos en el orden deseado
    data = []
    contador = 0
    for element in elements:
        valor = element.text
        if contador == 0:
            data.append({"Empresa": valor})
        elif contador == 1:
            data.append({"Puesto": valor})
        elif contador == 2:
            data.append({"Tiempo": valor})
        elif contador == 3:
            data.append({"Tipo": valor})
        elif contador == 4:
            data.append({"Pais": valor})
        elif contador == 5:
            data.append({"Ciudad": valor})
        elif contador == 6:
            data.append({"Rango Salarial": valor})
        elif contador == 7:
            data.append({"Requerimiento": valor})
        
        contador += 1
        if contador > 7:
            contador = 0

    # Crear el dataframe
    df = pd.DataFrame(data)
    print(df.head(10))

finally:
    # Cerrar el navegador
    driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Configurar Selenium
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chromedriver_path = "C:\Program Files (x86)\Google\chromedriver.exe"  # Reemplazar con la ruta del controlador ChromeDriver
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# URL de la página a hacer scraping
url = "https://jobrun.dev/ofertas"

# Acceder a la página
driver.get(url)

try:
    # Esperar a que aparezcan los contenedores de trabajos
    containers = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.__wab_flex-container div.plasmic_job_run_all__6OLuV.PlasmicOfferCard_freeBox___8Vji7__iE5Rg"))
    )

    data_jobs = []
    
    # Iterar sobre los contenedores y extraer los datos
    for container in containers:
        empresa = container.find_element(By.CSS_SELECTOR, "div.__wab_slot-string-wrapper").text
        puesto = container.find_element(By.CSS_SELECTOR, "div.plasmic_job_run_all__6OLuV.PlasmicOfferCard_freeBox__zEZtJ__h_k5H div.__wab_slot-string-wrapper").text
        salario = container.find_element(By.CSS_SELECTOR, "div.plasmic_job_run_all__6OLuV.PlasmicOfferCard_freeBox__wq10V__axvOC div.__wab_slot-string-wrapper").text
        lugar = container.find_element(By.CSS_SELECTOR, "div.plasmic_job_run_all__6OLuV.PlasmicOfferCardDetailsItem_svgtype_ciudad__FmyGa div.__wab_slot-string-wrapper").text
        tiempo = container.find_element(By.CSS_SELECTOR, "div.plasmic_job_run_all__6OLuV.PlasmicOfferCardDetailsItem_svgtype_remote__hPyK9 div.__wab_slot-string-wrapper").text
        
        data_jobs.append({
            "Empresa": empresa,
            "Puesto": puesto,
            "Salario": salario,
            "Lugar": lugar,
            "Tiempo": tiempo
        })

    # Crear el dataframe
    df = pd.DataFrame(data_jobs)
    
    # Guardar el dataframe en un archivo CSV
    df.to_csv("jobs_data.csv", index=False)

    # Imprimir las 10 primeras filas
    print(df.head(10))

finally:
    # Cerrar el navegador
    driver.quit()
```
