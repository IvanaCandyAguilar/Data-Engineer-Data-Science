```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

try:
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Ejecutar en modo headless (sin interfaz gráfica)
    chrome_options.add_argument("--no-sandbox")  # Opción adicional para evitar errores en algunos sistemas

    # Crear una instancia del controlador Chrome WebDriver en modo oculto
    driver = webdriver.Chrome(options=chrome_options)

    # Abrir el sitio web de Wikipedia
    driver.get("https://es.wikipedia.org")

    # Esperar a que el formulario de búsqueda esté presente en la página
    search_form = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.NAME, "search")))

    # Esperar a que el formulario de búsqueda sea interactuable
    search_form = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.NAME, "search")))

    # Ingresar la búsqueda y enviar el formulario
    search_form.send_keys("universidad España")
    search_form.submit()

    # Esperar a que los resultados estén presentes en la página
    results = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "mw-search-results")))

    # Imprimir el texto del primer resultado
    print(results[0].text)

    # Cerrar el navegador
    driver.quit()

except TimeoutException:
    print("No se encontraron elementos con la clase 'result' dentro del tiempo especificado.")
```
