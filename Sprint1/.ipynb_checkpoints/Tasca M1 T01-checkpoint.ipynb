### TABLA MOVIE

Esta tabla nos muestra los datos caracerísticos de una película (movie): Titulo, fecha de creación o lanzamiento (movie_date), formato (movie_format), un id que se relaciona con la tabla genero (movie_genre_id), usuario (created_by_user), fecha(created_date), fecha acualizacion(updated_date).

### TABLA GENRE

Contiene datos de los generos o catalogación de cada pelicula, pudiendo ser esta de tipo accion, comedia, romance, etc., los datos principales que se guardan son: id (genre_id), nombre de genero(genre_name), usuario quien creo o ingreso el dato(created_date), fecha de ingreso del dato(updated_date), fecha de actualizacion(update_date).

## EJERCICIO 4

Obtener, para cada persona, su nombre y el número máximo de roles distintos que ha tenido en una misma película.

```python
import mysql.connector

import pandas as pd

bd_movies = mysql.connector.connect(
  host="localhost",
  user="root",
  password="ContraseñaSegura123+",
  database="movies"
)

cursor_bd = bd_movies.cursor()

sentencia = 'SELECT PER.person_name, MOV.movie_title, count(MOV_PER.role_id) FROM tb_person PER INNER JOIN tb_movie_person MOV_PER ON PER.person_id = MOV_PER.person_id INNER JOIN tb_movie MOV ON MOV.movie_id = MOV_PER.movie_id GROUP BY MOV_PER.person_id, MOV_PER.movie_id ORDER BY PER.person_name'

cursor_bd.execute(sentencia)

resultado = cursor_bd.fetchall()

df = pd.DataFrame(resultado)

print (df)
```

Posteriormente, mostrar únicamente a aquellas personas que hayan asumido más de un rol en una misma película.

```python
import mysql.connector

bd_movies = mysql.connector.connect(
  host="localhost",
  user="root",
  password="ContraseñaSegura123+",
  database="movies"
)

cursor_bd = bd_movies.cursor()

sentencia = 'SELECT PER.person_name, MOV.movie_title, count(MOV_PER.role_id) AS num_rol FROM movies.tb_person PER INNER JOIN movies.tb_movie_person MOV_PER ON PER.person_id = MOV_PER.person_id INNER JOIN movies.tb_movie MOV ON MOV.movie_id = MOV_PER.movie_id GROUP BY MOV_PER.person_id, MOV_PER.movie_id HAVING num_rol > 1 ORDER BY PER.person_name'

cursor_bd.execute(sentencia)

resultado = cursor_bd.fetchall()

for x in resultado:
  print(x)
```
