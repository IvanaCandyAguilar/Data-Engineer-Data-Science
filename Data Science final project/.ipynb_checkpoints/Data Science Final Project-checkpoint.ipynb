```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

# Cargar el dataset preprocesado

df_pred = df_main_preprocesado

# Convertir las columnas "Ano_Inicial" y "Ano_Final" a tipo entero
df_pred['Ano_Inicial'] = df_pred['Ano_Inicial'].astype(int)
df_pred['Ano_Final'] = df_pred['Ano_Final'].astype(int)

# Crear el DataFrame df_result vacío
df_result = pd.DataFrame(columns=['Ano_Inicial', 'Ano_Final', 'Codi_Districte', 'Codi_Barri', 'Nombre_prediccion'])

# Recorrer los distritos y barrios únicos en el dataset
distritos_barrios = df_pred[['Codi_Districte', 'Codi_Barri']].drop_duplicates()
for _, row in distritos_barrios.iterrows():
    codi_districte = row['Codi_Districte']
    codi_barri = row['Codi_Barri']
    
    # Filtrar los datos para el distrito y barrio actual
    df_pred_districte_barri = df_pred[(df_pred['Codi_Districte'] == codi_districte) & (df_pred['Codi_Barri'] == codi_barri)]
    
    # Filtrar los datos para años anteriores al 2020
    df_pred_anteriores = df_pred_districte_barri[(df_pred_districte_barri['Ano_Inicial'] < 2020) & (df_pred_districte_barri['Ano_Final'] < 2024)]
    
    if len(df_pred_anteriores) > 0:
        # Dividir los datos en variables independientes (X_pred) y variable dependiente (y_pred)
        X_pred = df_pred_anteriores[['Ano_Inicial', 'Ano_Final']]
        y_pred = df_pred_anteriores['Nombre']
        
        # Generar características polinomiales de grado 2
        poly = PolynomialFeatures(degree=2)
        X_pred_poly = poly.fit_transform(X_pred)
        
        # Crear y entrenar el modelo de regresión lineal
        model = LinearRegression()
        model.fit(X_pred_poly, y_pred)
        
        # Realizar la predicción para el año inicial 2020 y final 2024 utilizando validación cruzada
        X_pred_prediccion = pd.DataFrame({'Ano_Inicial': [2020], 'Ano_Final': [2024]})
        X_pred_prediccion_poly = poly.transform(X_pred_prediccion)
        y_pred_prediccion = cross_val_score(model, X_pred_poly, y_pred, cv=5).mean()
        
        # Agregar los resultados al DataFrame df_result
        df_result = pd.concat([df_result, pd.DataFrame({
            'Ano_Inicial': [2020],
            'Ano_Final': [2024],
            'Codi_Districte': [codi_districte],
            'Codi_Barri': [codi_barri],
            'Nombre_prediccion': y_pred_prediccion
        })], ignore_index=True)

# Imprimir el DataFrame df_result
print(df_result)
```
