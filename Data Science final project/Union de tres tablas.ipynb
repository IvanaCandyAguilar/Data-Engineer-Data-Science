```python
import pandas as pd

df_taza_fecundidad_adolescentes = pd.read_csv('2020_est_salut_publica_tax_espec_fec_mares_15_19a.csv')
df_ultima_gestion = df_taza_fecundidad_adolescentes[df_taza_fecundidad_adolescentes['Anys_Periode'] == '2016-2020']
df_media_cantidad_adolescente = df_ultima_gestion.groupby("Nom_Barri")["Nombre"].mean().reset_index()
df_media_por_barri
```

```python
df_renta_media = pd.read_csv('2020_atles_renda_mitjana.csv')
df_media_ingresos_por_barrio = df_renta_media.groupby('Nom_Barri')['Mitjana_Renda_€'].mean().reset_index()
df_media_ingresos_por_barrio_ordenado = df_media_ingresos_por_barrio.sort_values('Mitjana_Renda_€',ascending=False)
#df_media_ingresos_por_barrio_ordenado
```

```python
df_centro_deportivos = pd.read_csv('opendatabcn_esports_esports-individuals.csv', encoding='UTF-16')
df_numero_centros_deportivos_por_barrio = df_centro_deportivos.groupby('addresses_neighborhood_name').size().reset_index(name='Numero_de_Centros_Deportivos')
#df_numero_centros_deportivos_por_barrio
```

```python
df_locales_nocturnos = pd.read_csv('opendatabcn_cultura_espais-de-musica-i-copes.csv', encoding='UTF-16')
df_numero_locales_nocturnos_por_barrio = df_locales_nocturnos.groupby('addresses_neighborhood_name').size().reset_index(name='Numero_de_Centros_Nocturnos')
#df_numero_locales_nocturnos_por_barrio
```

```python
# Unir el primer par de DataFrames
df_unido_1 = pd.merge(df_media_cantidad_adolescente, df_media_ingresos_por_barrio_ordenado, on='Nom_Barri', how='outer')

#df_unido_1

# Unir el segundo par de DataFrames
df_unido_2 = pd.merge(df_numero_centros_deportivos_por_barrio, df_numero_locales_nocturnos_por_barrio, on='addresses_neighborhood_name', how='outer')
df_unido_2 = df_unido_2.rename(columns={'addresses_neighborhood_name': 'Nom_Barri'})
#df_unido_2
#df_unido = pd.merge(df_media_ingresos_por_barrio_ordenado, df_numero_centros_deportivos_por_barrio, left_on='Nom_Barri', right_on='addresses_neighborhood_name', how='inner')

# Unir los dos resultados anteriores
df_final = pd.merge(df_unido_1, df_unido_2, on='Nom_Barri', how='outer')
df_final_ordenado = df_final.sort_values('Numero_de_Centros_Nocturnos',ascending=True)
df_final_ordenado.head(50)
# Seleccionar las columnas deseadas
#df_final = df_final[['nombre_barri', 'media_cantidad_adolescente', 'media_ingresos', 'cantidad_centros_deportivos', 'locales_nocturnos']]
```
